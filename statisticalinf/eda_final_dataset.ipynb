```python
import pandas as pd
import numpy as np
from IPython.display import display


df = pd.read_csv('g4m.csv')


print("Dataset Shape:", df.shape)
display(df.head())
df.info()
display(df.describe(include='all'))


print("Missing values per column:")
print(df.isnull().sum().sort_values(ascending=False).head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot count of target variable 'Bleaching'
plt.figure(figsize=(6,4))
sns.countplot(x='Bleaching', data=df)
plt.title('Bleaching Counts')
plt.show()

# Pie chart of Bleaching
plt.figure(figsize=(6,6))
plt.pie(df['Bleaching'].value_counts(), labels=['No Bleaching','Bleaching'], 
        autopct='%1.1f%%', colors=['#66b3ff','#ff6666'])
plt.title('Bleaching Proportion')
plt.show()

# Print class balance
bleaching_counts = df['Bleaching'].value_counts(normalize=True)
print("Bleaching Class Distribution (fraction):")
print(bleaching_counts)
```

```python
# List of numerical features for correlation
numerical_features = ['Year','fCO2','pH_total','SST@90th_HS','SSTA@90th_HS','SST_MAX','SST_MIN','Salinity','DHW_from_90th_HS>1','ENSO','IOD']

# Compute correlation matrix for numerical features
corr_matrix = df[numerical_features].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Cell 6: Correlation vs Bleaching
target_corr = df['Bleaching'].sort_values(ascending=False)
display(target_corr)
```

```python
# Histograms for numerical features
for col in numerical_features:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
# Boxplots of numerical features vs target
for col in numerical_features:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Bleaching', y=col, data=df)
    plt.title(f'{col} by Bleaching')
    plt.show()

# Countplots for selected categorical/binary features vs Bleaching
selected_cats = ['Acropora', 'Acanthastrea', 'Stylophora']
for col in selected_cats:
    if col in df.columns:
        plt.figure(figsize=(6,4))
        sns.countplot(x=col, hue='Bleaching', data=df)
        plt.title(f'{col} presence vs Bleaching')
        plt.show()
```

```python
# Pairplot of key variables with hue=Bleaching
pairplot_features = ['Year','fCO2','pH_total','Salinity','SST_MAX','Bleaching']
sns.pairplot(df[pairplot_features], hue='Bleaching', diag_kind='hist')
plt.suptitle('Pairplot of Key Variables', y=1.02)
plt.show()

# Stacked bar chart: Bleaching counts per Year
bleach_by_year = pd.crosstab(df['Year'], df['Bleaching'])
bleach_by_year.plot(kind='bar', stacked=True, figsize=(8,6), color=['#66b3ff','#ff6666'])
plt.title('Bleaching by Year')
plt.ylabel('Count')
plt.legend(title='Bleaching')
plt.show()

# Swarmplot of a numerical feature vs Bleaching
plt.figure(figsize=(6,4))
sns.swarmplot(x='Bleaching', y='Salinity', data=df)
plt.title('Salinity by Bleaching')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Prepare data for model
X = df.drop(columns=['Bleaching'])
y = df['Bleaching']

# Fit a simple Random Forest to get feature importances
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)
importances = pd.Series(rf.feature_importances_, index=X.columns)
top_features = importances.sort_values(ascending=False).head(10)

plt.figure(figsize=(8,6))
sns.barplot(x=top_features.values, y=top_features.index, palette='viridis')
plt.title('Top 10 Feature Importances (Random Forest)')
plt.xlabel('Importance')
plt.show()
```
