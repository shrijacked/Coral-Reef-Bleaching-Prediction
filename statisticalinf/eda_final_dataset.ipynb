```python
import pandas as pd
import numpy as np
from IPython.display import display


df = pd.read_csv('g4m.csv')


print("Dataset Shape:", df.shape)
display(df.head())
df.info()
display(df.describe(include='all'))


print("Missing values per column:")
print(df.isnull().sum().sort_values(ascending=False).head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot count of target variable 'Bleaching'
plt.figure(figsize=(6,4))
sns.countplot(x='Bleaching', data=df)
plt.title('Bleaching Counts')
plt.show()

# Pie chart of Bleaching
plt.figure(figsize=(6,6))
plt.pie(df['Bleaching'].value_counts(), labels=['No Bleaching','Bleaching'], 
        autopct='%1.1f%%', colors=['#66b3ff','#ff6666'])
plt.title('Bleaching Proportion')
plt.show()

# Print class balance
bleaching_counts = df['Bleaching'].value_counts(normalize=True)
print("Bleaching Class Distribution (fraction):")
print(bleaching_counts)
```
